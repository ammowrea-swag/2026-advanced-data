````python
# -----------------------------
# 1) Load and clean the dataset
# -----------------------------
# Update path if needed
df = pd.read_csv("summons-15-25.csv")

# Replace common missing-value markers with pandas missing value
df = df.replace({"": pd.NA, "(null)": pd.NA, "UNKNOWN": pd.NA})

# Normalize borough names
df["BORO_CLEAN"] = df["BORO"].str.upper()

# Ensure YEAR is numeric
df["YEAR"] = pd.to_numeric(df["YEAR"], errors="coerce")

# Clean sex values: keep only M/F, everything else -> missing
df["SEX_CLEAN"] = df["SEX"].str.upper()
df["SEX_CLEAN"] = df["SEX_CLEAN"].where(df["SEX_CLEAN"].isin(["M", "F"]), pd.NA)

# Clean race
df["RACE_CLEAN"] = df["RACE"].str.upper()
````

````python
# ---------------------------------------
# Q1) Which borough had the most summons?
# ---------------------------------------
summons_by_boro = df["BORO_CLEAN"].value_counts().rename("summons_count")

top_borough = summons_by_boro.idxmax()
top_count = summons_by_boro.max()

print(f"Q1: {top_borough} had the most summons with {top_count:,} total.")
````

````python
# ----------------------------------------------------------------
# Q2) How many times did the top 10 most common offenses occur?
# ----------------------------------------------------------------

top_10_offenses = df["OFFENSE_DESCRIPTION"].value_counts().nlargest(10)

top_offense = top_10_offenses.idxmax()
top_offense_count = top_10_offenses.max()
top_10_total = top_10_offenses.sum()

print(
    f"\nQ2: The top 10 most common offenses occurred {top_10_total:,} times combined. "
    f"The #1 offense was '{top_offense}' with {top_offense_count:,} occurrences.\n"
    f"{top_10_offenses.to_string()}"
)
````

````python
# ------------------------------------------------------
# Q3) Total number of summons broken down by year
# ------------------------------------------------------
summons_by_year = (
    df.dropna(subset=["YEAR"])
      .groupby(df["YEAR"].astype(int))
      .size()
      .rename("summons_count")
      .sort_index()
)

peak_year = summons_by_year.idxmax()
peak_count = summons_by_year.max()

print(
    f"\nQ3: {summons_by_year.sum():,} total summons were issued. "
    f"The peak year was {peak_year} with {peak_count:,} summons.\n"
    f"{summons_by_year.to_string()}"
)
````

````python

# -----------------------------------------------------------------------------------
# Q4) Most unusual summons (rarest offense descriptions)
# -----------------------------------------------------------------------------------
offense_counts = df["OFFENSE_DESCRIPTION"].value_counts()

singletons = offense_counts[offense_counts == 1]
very_rare   = offense_counts[offense_counts <= 2].sort_values()

print(
    f"\nQ4: {len(singletons)} offense(s) appeared exactly once (rarest). "
    f"{len(very_rare)} offense(s) had 2 or fewer occurrences.\n"
    f"Singleton offenses:\n{singletons.to_string()}"
)
````

````python
# -------------------------------------
# Q5) Count of summons by gender
# -------------------------------------
summons_by_gender = (
    df["SEX_CLEAN"]
    .value_counts()
    .rename(index={"M": "Male", "F": "Female"})
)

print(
    f"\nQ5: Summons by gender — "
    + ", ".join(f"{g}: {c:,}" for g, c in summons_by_gender.items())
)
````

````python
# --------------------------------------------
# Q6) How many summons were issued by race?
# --------------------------------------------
summons_by_race = df["RACE_CLEAN"].value_counts().rename("summons_count")

top_race = summons_by_race.idxmax()
top_race_count = summons_by_race.max()

print(
    f"\nQ6: {summons_by_race.sum():,} summons had a recorded race. "
    f"The most common was {top_race} with {top_race_count:,} summons.\n"
    f"{summons_by_race.to_string()}"
)
````

````python
# ------------------------------------------------------------------------
# Q7) Which borough had the fewest offenses per 100K people?
# ------------------------------------------------------------------------
borough_population = pd.Series({
    "BRONX":         1_472_654,
    "BROOKLYN":      2_736_074,
    "MANHATTAN":     1_694_251,
    "QUEENS":        2_405_464,
    "STATEN ISLAND":   495_747,
}, name="population")

rates_df = (
    df["BORO_CLEAN"]
    .value_counts()
    .rename("summons_count")
    .to_frame()
    .join(borough_population)
    .assign(offenses_per_100k=lambda x: (x["summons_count"] / x["population"]) * 100_000)
    .sort_values("offenses_per_100k")
)

fewest_boro  = rates_df.index[0]
fewest_rate  = rates_df["offenses_per_100k"].iloc[0]
highest_boro = rates_df.index[-1]
highest_rate = rates_df["offenses_per_100k"].iloc[-1]

print(
    f"\nQ7: {fewest_boro} had the fewest offenses per 100K residents ({fewest_rate:.2f}), "
    f"while {highest_boro} had the highest ({highest_rate:.2f}).\n"
    f"{rates_df.to_string()}"
)
````
